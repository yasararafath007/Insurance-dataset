Data preprocessing

```python
data['Gender'] = OneHotEncoder().fit_transform(data[['Gender']]).toarray()
data['Vehicle_Age'] = OneHotEncoder().fit_transform(data[['Vehicle_Age']]).toarray()
data['Vehicle_Damage'] = OneHotEncoder().fit_transform(data[['Vehicle_Damage']]).toarray()

dataa['Gender'] = OneHotEncoder().fit_transform(dataa[['Gender']]).toarray()
dataa['Vehicle_Age'] = OneHotEncoder().fit_transform(dataa[['Vehicle_Age']]).toarray()
dataa['Vehicle_Damage'] = OneHotEncoder().fit_transform(dataa[['Vehicle_Damage']]).toarray()

train = data.drop(columns="id")
test = dataa.drop(columns="id")
```

```python
x = train.drop(columns="Response")
y = train["Response"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Model building

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred))
```
